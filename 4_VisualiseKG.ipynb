```python
# graph visualization
from pyvis.network import Network
# show HTML in notebook
import IPython
import pandas as pd
```

```python
def excel_to_dict(file_path):

    df = pd.read_excel(file_path)

    dict_list = []

    for index, row in df.iterrows():
        relation_dict = {
            'chunk_id': row['chunk_id'],
            'head': row['head'],
            'type': row['type'],
            'tail': row['tail']
        }
        dict_list.append(relation_dict)

    return dict_list
```

```python
file_path = '/content/triples_data.xlsx'
relations = excel_to_dict(file_path)
for relation in relations:
    print(relation)
```

```python
from pyvis.network import Network

def save_network_html(relations, filename="network.html"):
    # Create network
    net = Network(directed=True, width="auto", height="700px", bgcolor="#333333")

    # Create a set to keep track of unique entities
    entities = set()

    # Iterate over the relations list to extract entities
    for r in relations:
        # Add entities to the set
        entities.add(r['head'])
        entities.add(r['tail'])

    # Add nodes to the network
    color_entity = "#40E0D0"
    for e in entities:
        net.add_node(e, shape="circle", color=color_entity)

    # Add edges
    for r in relations:
        net.add_edge(r['head'], r['tail'], title=r['type'], label=r['type'])

    # Save network
    net.repulsion(
        node_distance=200,
        central_gravity=0.2,
        spring_length=200,
        spring_strength=0.05,
        damping=0.09
    )
    net.set_edge_smooth('dynamic')
    net.show(filename)
```

```python
filename = "foodwise_and_foodvision.html"
save_network_html(relations, filename=filename)
```
